```python
df.isnull().sum()
```

```python
dummy_sex = pd.get_dummies(df['Sex'],drop_first=True)
dummy_sex
```

```python
Embarked_replace = pd.get_dummies(df['Embarked'],drop_first=True)
Embarked_replace.head()
```

```python
b = pd.get_dummies(df['Cabin_replaced'],drop_first=True)
b.head()
```

```python
df = pd.concat([df,dummy_sex,Embarked_replace,b],axis=1)
```

```python
df
```

```python
df.info()
```

```python
df = df.drop(['PassengerId','Name','Sex','Ticket','Cabin','Cabin_replaced','Embarked'],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
X = df.drop(['Survived'],axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
learning_rates = [0.05,0.1,0.02,0.001,0.01,0.25]
for i in learning_rates:
    gra = GradientBoostingClassifier(learning_rate=i,n_estimators=100,random_state=100)
    gra.fit(X_train,y_train)
    print("learning rates: ", i)
    print("Accuracy score (training): {0:.3f}".format(gra.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gra.score(X_test, y_test)))
    print()
```

```python
y_train_pred = gra.predict(X_train)
```

```python
accuracy_score(y_train,y_train_pred)*100
```

```python
y_test_pred = gra.predict(X_test)
```

```python
accuracy_score(y_test,y_test_pred)*100
```
